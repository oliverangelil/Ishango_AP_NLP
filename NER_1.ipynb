```python
def getorgEnt(doc):
    orgs = []
    for org in doc.ents:
        if org.label_ == "ORG":
            orgs.append(org)
    return orgs

def getNorpEnt(doc):
    norps = []
    for ent in doc.ents:
        if ent.label_ == "NORP":
            norps.append(ent)
    return norps
```
